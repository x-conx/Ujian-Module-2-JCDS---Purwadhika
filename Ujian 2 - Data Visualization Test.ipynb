```python
asd = country3.get_group('Asia')
asd.head()
```

```python
asd.LifeExpectancy.describe()
```

## IQR

```python
iqr = asd.LifeExpectancy.quantile(.75) - asd.LifeExpectancy.quantile(.25)
mean = asd.LifeExpectancy.mean()
lower_bound = asd.LifeExpectancy.quantile(.25) - (1.5*iqr)
upper_bound = asd.LifeExpectancy.quantile(.75) + (1.5*iqr)
median = asd.LifeExpectancy.median()
std = asd.LifeExpectancy.std()
print(f'IQR = {iqr}, mean={mean}, median={median}, lower bound={lower_bound}, upper bound={upper_bound}, std={std}')
```

```python
# fig.ax = plt.subplot(asd.LifeExpectancy)
sns.distplot(asd.LifeExpectancy)
plt.axvline(asd.LifeExpectancy.mean(),color='g',label='mean')
plt.axvline(asd.LifeExpectancy.median(),color='r',label='median')
plt.axvline(upper_bound,label='Upper_bound')
plt.axvline(lower_bound,color='black',label='lower_bound')
plt.legend(loc="upper left")
```
